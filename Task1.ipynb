```python
# create class Account
class Account:
    def __init__(self, account, pin, balance):   # __init__()method
        self.transaction_history = []
        self.account = account
        self.user_pin = pin
        self.balance = balance
       

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transaction_history.append(f'Amount Deposited: +${amount}')
            return f'Balance: ${self.balance}, Deposit: ${amount}'
        else:
            return "Deposit amount must be greater than 0."

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            self.transaction_history.append(f'Amount Withdrawn: -${amount}')
            return f'Balance: ${self.balance}, Withdrawn: ${amount}'
        else:
            return "Withdrawal amount is invalid."
        
        # display transaction history
    def display_transaction_history(self):
        return '\n'.join(self.transaction_history)

    def transfer(self, recipient_account_no, amount):
        if recipient_account_no in account_numbers:
            recipient_index = account_numbers.index(recipient_account_no)
            if amount > 0 and self.balance >= amount:
                self.balance -= amount
                balances[recipient_index] += amount
                self.transaction_history.append(f'Amount transferred to ${recipient_account_no}: ${amount}')
                return f'Balance: ${self.balance}'
            else:
                return "Transfer amount is invalid."
        else:
            return "Recipient account does not exist."

    def quit(self):
        return "Thank you... Have a Nice day!"

account_numbers = ["301220", "250602"]
pin_store = [301, 256]
balances = [1000, 2000]

account_no = input("Enter your acc. no.:")
account_pin = eval(input("Enter account pin:"))

# Find the index of the account number in the list
if account_no in account_numbers:
    account_index = account_numbers.index(account_no)
    account = Account(account_no, account_pin, balances[account_index])
else:
    print("Account not found or incorrect PIN.")
    exit()

def main():
    while True:
        print("\n***ATM***")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Transaction History")
        print("4. Transfer")
        print("5. Quit")
        choice = int(input("Enter your choice between 1-5:"))
        
        if choice == 1:
            amount = eval(input("Enter the amount to be deposited:"))
            print(account.deposit(amount))
        elif choice == 2:
            amount = eval(input("Enter the amount to be withdrawn:"))
            print(account.withdraw(amount))
        elif choice == 3:
            print(account.display_transaction_history())
        elif choice == 4:
            recipient_account_no = input("Enter the account no.For transferred money:")
            amount = eval(input("Enter the amount to be transferred:"))
            print(account.transfer(recipient_account_no, amount))
        elif choice == 5:
            print(account.quit())
            break
        else:
            print("INVALID CHOICE! SELECT BETWEEN 1-5.")

if __name__ == "__main__":
    main()
```
